```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rcParams["figure.facecolor"] = "w"
```

```python
data_home = "./"
```

```python
kokkos_data = np.loadtxt(f"{data_home}/kokkos_buffer_packing-timings.dat")
```

```python
fig,axes= plt.subplots()

ax = axes

ax.plot(kokkos_data[:,1],kokkos_data[:,7]/1e-6,label="Packing",marker="o")
ax.plot(kokkos_data[:,1],kokkos_data[:,10]/1e-6,label="Unpacking",marker="o")

ax.set_xlabel("Grid size")
ax.set_ylabel("Kernel time [$\\mu s$]")

ax.set_xscale("log")
ax.set_yscale("log")

ax.set_xticks(kokkos_data[:,1].astype(int),minor=False)
ax.set_xticklabels([ f"${int(nx):0d}^3$" for nx in kokkos_data[:,1]],minor=False)

ax.legend()

plt.show()
```
